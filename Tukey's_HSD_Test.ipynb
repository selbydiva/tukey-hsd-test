```python
import pandas as pd
import numpy as np
```

```python
data = {
    'id_movie' : ['A1','A2','A3','A4','A5', 'A6', 'A7', 'A8', 'A9'],
    'genre' : ['Romance' , 'Horror', 'Comedy', 'Comedy', 'Horror', 'Romance', 'Comedy', 'Horror', 'Romance'],
    'rating' : [3.5, 3.2, 4, 3.8, 3.6, 3.8, 4.1, 3.5, 3.7]
}

data = pd.DataFrame(data)
data
```

```python
from statsmodels.formula.api import ols
import pandas as pd
import statsmodels.api as sm

df = data[['rating','genre']]

# Fit OLS model
model = ols('rating ~ C(genre)', data=df).fit()

# Perform ANOVA on the model
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
import scipy.stats as stats

# Parameters
alpha = 0.95
dof_between = 2
dof_within = 6

# Calculate the F-distribution critical value
critical_value = stats.f.ppf(alpha, dof_between, dof_within)

print(f"The critical F-value at alpha = {alpha} with {dof_between} and {dof_within} degrees of freedom is: {critical_value}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey = pairwise_tukeyhsd(endog=df['rating'], groups=df['genre'], alpha=0.05)
print(tukey)
```

```python
import statsmodels.stats.libqsturng as qsturng

# Example: Find the critical value for k=4 groups, dfw=20 degrees of freedom, alpha=0.05
k = 3
dfw = 6
alpha = 0.05

# Get the critical value
critical_value = qsturng.qsturng(1 - alpha, k, dfw)
print(f"Critical value for alpha={alpha}, k={k}, dfw={dfw}: {critical_value}")
```
